### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

#Promedio
suma = 0
for esperanza in esperanzas_de_vida:
    suma += esperanza
promedio = suma / len(esperanzas_de_vida)
print(f"Promedio de esperanza de vida: {promedio:.2f} años")

#Máximo
maximo = esperanzas_de_vida[0]
for esperanza in esperanzas_de_vida:
    if esperanza > maximo:
        maximo = esperanza
print(f"Valor máximo de esperanza de vida: {maximo:.2f} años")

#Mínimo
minimo = esperanzas_de_vida[0]
for esperanza in esperanzas_de_vida:
    if esperanza < minimo:
        minimo = esperanza
print(f"Valor mínimo de esperanza de vida: {minimo:.2f} años")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí

def clasificar_esperanza_vida(años):
    if años>= 75:
        return "Alta"
    elif 60 <= años < 75:
        return "Media"
    else:
        return "Baja"
    
#Lista de valores para probar la función
valores_prueba = [78.6, 73.4, 38.0, 90.0, 54.9, 65.0]

for valor in valores_prueba:
    resultado = clasificar_esperanza_vida(valor)
    print(f"Esperanza de vida: {valor} años - Clasificación: {resultado}")
````

### 1.4 — Listas + Funciones + Bucles (integrador)

**Ejercicio 1.4** — Usa la función creada en 1.3 y la lista del ejercicio 1.1 para:
1. Crear una **nueva lista** llamada `clasificaciones` que contenga la clasificación de cada esperanza de vida
2. Imprimir cada par: `(valor, clasificación)` usando un bucle `for`

````python
# Tu código aquí

#En una variable se guarda los daros de esperanzas de vida
clasificaciones = [clasificar_esperanza_vida(años) for años in esperanzas_de_vida]

#Se imprime el valor y su clasificación en la lista usando la función
for años, clasificacion in zip(esperanzas_de_vida, clasificaciones):
    print(f"Esperanza de vida: {años} años - Clasificación: {clasificacion}")
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
#Primero reaalizamos una descripción estadística por columna con el método describe()
#para obtener una visión general de los datos.
print(f"Descripción estadística:\n{df.describe()}")

#Se reemplazan los valores nulos por la media de cada columna

#Se crea una lista con los nombres de las columnas a las que se les va a reemplazar los valores nulos por la media
CambiarColumnas = ['esperanza_vida', 'gasto_salud_usd', 'mortalidad_infantil', 
    'medicos_por_1000', 'poblacion_urbana_pct', 'obesidad_pct', 
    'diabetes_pct', 'tabaquismo_pct', 'camas_hospital_por_1000', 
    'vacunacion_pct']

#Se itera sobre la lista de columnas y se reemplazan los valores nulos por la media de cada columna utilizando el método fillna() de pandas
for columna in CambiarColumnas:
    df[columna] = df[columna].fillna(df[columna].mean())

#Se verifica que ya no hay valores nulos en el dataframe
df.isnull().sum()
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: 
Se reemplazan los valores nulos por la media, ya que este valor representa el estándar
de cada columna y no afecta significativamente el dataframe. En cambio, si se eliminan
filas, esto podría tener un impacto considerable en el dataframe.

### 2.3 — Estadísticas descriptivas

3. ¿Cuál es el país con **mayor** gasto en salud? ¿Y el de **menor**?
4. ¿Cuál es la **mediana** de la mortalidad infantil?

### 2.4 — Filtrado y selección

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
#Se calcula la esperanza de vida por region
esperanza_vida_region = df.groupby('region')['esperanza_vida'].mean()
print(f"\nEsperanza de vida por región:\n{esperanza_vida_region}")

#Se calcula los gastos de sald por nivel de ingresos
gastoSalud_ingresos = df.groupby('nivel_ingresos')['gasto_salud_usd'].mean()
print(f"\nGastos de salud por nivel de ingresos:\n{gastoSalud_ingresos}")

#Se calcula el numero de paises por region
paises_region = df.groupby('region')['pais'].count()
print(f"\nNúmero de países por región:\n{paises_region}")

#Se filtra las regiones con mayor y menor esperanza de vida promedio
mayor_esperanza = df.groupby('region')['esperanza_vida'].idxmax()
menor_esperanza = df.groupby('region')['esperanza_vida'].idxmin()
print(f"\nPaíses con mayor esperanza de vida por región:\n{df.loc[mayor_esperanza]}")
print(f"\nPaíses con menor esperanza de vida por región:\n{df.loc[menor_esperanza]}")
````

🔧 **Ejercicio 2.6**

````python
# Tu código aquí
#Se crea la columna 'categoria_vida' utilizando la función clasificar_esperanza_vida
df['categoria_vida'] = df['esperanza_vida'].apply(clasificar_esperanza_vida)
print(f"\nDataFrame con la nueva columna 'categoria_vida':\n{df[['esperanza_vida', 'categoria_vida']].head()}")

#Se cuenta el número de países en cada categoría
categoria_counts = df['categoria_vida'].value_counts()
print(f"\nNúmero de países en cada categoría de esperanza de vida:\n{categoria_counts}")

#Se muestra el porcentaje de países en cada categoría de esperanza de vida
categoria_percentages = df['categoria_vida'].value_counts(normalize=True) * 100
print(f"\nPorcentaje de países en cada categoría de esperanza de vida:\n{categoria_percentages}")
````
